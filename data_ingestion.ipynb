```python
import requests
import pandas as pd
import os
from datetime import datetime
import time
import pytz
```

```python
def fetch_bikepoint_data():
    url = "https://api.tfl.gov.uk/BikePoint"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else: 
        print(f"Failed to fetch data: {response.status_code}")
        return None
```

```python
def save_data(data, directory):
    gmt_time = datetime.now(pytz.timezone('GMT'))
    filename = f"/home/osbdet/notebooks/mda2/group_assignment/bikepoint_data/{gmt_time.strftime('%Y-%m-%d_%H%M%S')}"
    full_path = os.path.join(directory, filename)
    
    rows = []
    for point in data:
        for prop in point['additionalProperties']:
            if prop['key'] == 'NbBikes':
                rows.append({
                    'id': point['id'],
                    'name': point['commonName'],
                    'lat': point['lat'],
                    'lon': point['lon'],
                    'nbBikes': prop['value'],
                    'timestamp': gmt_time
                })
    df = pd.DataFrame(rows)
    
    df.to_csv(full_path, index=False, header=True)
    print(f"Data saved to {full_path}")
    
directory = os.path.join(os.path.expanduser('~'), 'data')

if not os.path.exists(directory):
    os.makedirs(directory)
    
while True:
    start_time = time.time()
    data = fetch_bikepoint_data()
    if data:
        save_data(data, directory)
    else:
        print("Failed to fetch data.")
        
    elapsed_time = time.time() - start_time
    wait_time = max(600 - elapsed_time, 0)
    print(f"Waiting for {wait_time} seconds until next cycle.")
    time.sleep(wait_time)
```
